```python
#Солдат и бананы 

first_b, soldiers_money, how_many = map(int, input().split()) 
cost = 0
for i in range(1, how_many+1):
    banana_cost = i * first_b
    cost += banana_cost
if soldiers_money >= cost:
    print(0)
else:
    print(cost - soldiers_money)
```

```python
#задача о делимости
t = int(input())

for _ in range(t):
    a, b = map(int, input().split())
    if a % b == 0:
        print(0)
    else:
        print(b - (a % b))
```

```python
#средство от бессоницы 
k = int(input())
l = int(input())
m = int(input())
n = int(input())
d = int(input())

count = 0

for i in range(1, d + 1):
    if i % k == 0 or i % l == 0 or i % m == 0 or i % n == 0:
        count += 1

print(count)
```

```python
#конфеты и две сестры
t = int(input())
for _ in range(t):
    n = int(input())
    ways = (n - 1) // 2
    if ways < 0:
        ways = 0
    print(ways)
```

```python
#новый год и спешка

n, k = map(int, input().split())

max_tasks = 0
time_needed = 0

for i in range(1, n+1):
    time_needed += 5*i
    if time_needed + k <= 240:
        max_tasks = i
    else:
        break

print(max_tasks)
```
